# How to visualize data i.e plotting

```python
tmp4 = np.random.randint(1,50,100).reshape(10,10)
df4 = pd.DataFrame(tmp4)
df4
```

```python
tmp5 = np.random.randint(1,50,100).reshape(10,10)
tmp5
```

### Line plot

```python
plt.plot(range(10), tmp5)
plt.show()
```

### Enhanced graph

```python
plt.figure(figsize=(16,4))
plt.plot(range(10),tmp5)
plt.show()
```

```python
plt.figure(figsize=(16,4))
plt.bar(range(10),tmp5)
plt.show()
```
